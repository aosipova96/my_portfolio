### Часть 2. Предобработка данных.

<a id='chapter_4'></a>Сначала изменим названия некоторых столбцов на более правильные.

```python
# переименуем столбцы с помощью метода rename()
data = data.rename(columns={'cityCenters_nearest':'city_centers_nearest', 
                            'parks_around3000':'parks_around_3000', 
                            'ponds_around3000':'ponds_around_3000'})
```

<a id='chapter_5'></a> Теперь посмотрим на неявные дубликаты в названиях населенных пунктов, т.к. это может помешать при удалении пропусков в данных. Выведем все уникальные значения столбца **locality_name**.

```python
# выведем все уникальные значения столбца locality_name
data['locality_name'].unique() 
```

Можно заметить, что есть неявные дубликаты в словах *'посёлок'* и *'поселок'*. Их достаточно много. Заменим все буквы *'ё'* на *'е'*.

```python
# заменяем букву ё на букву е в столбце locality_name
data['locality_name'] = data['locality_name'].str.replace('ё', 'е')
```

Также нужно удалить все неявные дубликаты по типу *'поселок городского типа Рябово'* и *'поселок Рябово'*. Значений достаточно много и искать их все вручную тяжело. Поэтому просто избавимся от типов населенных пунктов, такие как деревня, поселок, городской поселок и т.д.

```python
# создадим список с наименованиями типов населенных пунктов
list_of_locality = ['поселок городского типа ', 'городской поселок ', 'коттеджный поселок ', 
                    'садовое товарищество ', 'поселок при железнодорожной станции ',
                    'садоводческое некоммерческое товарищество ', 'поселок ', 'село ', 'деревня ']

# передадим этот список как параметр методу replace и заменим эти значения в столбце locality_name на пустую строку
data['locality_name'] = data['locality_name'].replace(list_of_locality, '', regex=True)
```

<a id='chapter_6'></a>Проверим датасет на явные дубликаты (полностью повторяющиеся строки).

```python
# ищем число строк, для которых метод duplicated() возвращает True
data.duplicated().sum()
```

<a id='chapter_7'></a>Полностью повторяющихся строк нет. Теперь найдем в каких столбцах имеются пропуски и сколько там таких пропусков.

```python
#найдем сумму всех пропусков в таблице
data.isna().sum()
```

Многие столбцы имеют пропуски в данных, причем достаточно весомые. Так, например, столбец **is_apartment** имеет 20924 пустых строки из 23699. Также много пропусков в столбцах **park_nearest** и **ponds_nearest**. Пропуски в столбце **balcony** можно заменить на нули, т.к. скорее всего продавец не указал их наличие, т.к. их нет в квартире.

```python
# присваиваем текущим значениям столбца новые значения с учетом заполненных пропусков
data['balcony'] = data['balcony'].fillna(0)

# выводим количество пустых значений в столбце
data['balcony'].isna().sum()
```

Столбец **balcony** имеет тип данных *float64*, но количество балконов не может быть дробным числом, поэтому изменим тип столбца на int64.

```python
# перезаписываем значения столбца balcony на значения с измененным типом
data['balcony'] = data['balcony'].astype('int64')
```

Рассмотрим столбец **is_apartment**. Посмотрим сколько всего уникальных значений есть в этом столбце.

```python
# выводим уникальные непустые значения столбца и количество таких значений
data['is_apartment'].value_counts(dropna=False)
```

Видим, что из почти 24 тысяч записей *непустое* значение имеют лишь **2775** строк. С формальной точки зрения апартаменты — это нежилые, но по характеристикам и удобствам практически не уступающие квартирам помещения, в которых можно жить. Основным отличием является то, что в них нельзя зарегистрироваться. Апартаменты так же могут иметь несколько комнат, балкон и находиться как в центре, так и вне его. Поэтому нет смысла искать взаимосвязь этого параметра с другими параметрами в таблице для заполнения пропусков. Скорее всего, в этом столбце пропуск равносилен **False**, т.е. квартира не является апартаментами. Заполним пропуски.

```python
# перезаписываем значения столбца на значения с заполненными пропусками
data['is_apartment'] = data['is_apartment'].fillna(False)
```

Проверим, что все пропуски заполнены.

```python
# ищем количество пропусков в столбце is_apartment
data['is_apartment'].isna().sum()
```

Столбец **is_apartment** имеет тип *object*, изменим его на булев тип, т.к. значениями столбца являются только True и False.

```python
# меняем тип данных на bool
data['is_apartment'] = data['is_apartment'].astype('bool')
```

Рассмотрим столбцы **parks_around_3000** и **ponds_around_3000**. Данные столбцы имеют по 5518 пропусков. Вероятнее всего, большинство пропусков в этих столбцах связано с тем, что квартиры находятся в маленьких пригородных городах или поселках, в которых нет ни прудов, ни парков. Получим список с названиями населенных пунктов, в которых для столбцов **parks_around_3000** и **ponds_around_3000** отсутствуют значения. 

```python
# находим срез датафрейма с условием, что в столбцах parks_around_3000 или ponds_around_3000 есть пропуски
(data.query('parks_around_3000.isna() == True or ponds_around_3000.isna() == True')['locality_name']
     .value_counts() # находим список уникальных значений в столбце locality_name
     .head(10))
```

Видим, что, в основном, это, действительно, небольшие города и поселки в Ленинградской области. Будем считать, что на расстоянии 3 км вблизи продаваемых квартир нет парков или прудов. Заменим пропущенные значения в этих столбцах на нули.

```python
# перезаписываем значения столбцов parks_around_3000 и ponds_around_3000 на значения с нулями вместо пропусков
data['parks_around_3000'] = data['parks_around_3000'].fillna(0)

data['ponds_around_3000'] = data['ponds_around_3000'].fillna(0)
```

Проверим, что пропусков в этих столбцах больше нет.

```python
# суммируем пропущенные значения в столбце parks_around_3000
print(f'Сумма пропусков в столбце parks_around_3000 - {data["parks_around_3000"].isna().sum()}')

# суммируем пропущенные значения в столбце ponds_around_3000
print(f'Сумма пропусков в столбце ponds_around_3000 - {data["ponds_around_3000"].isna().sum()}')
```

Теперь изменим тип значений этих столбцов на целочисленный, т.к. количество прудов или парков не может быть вещественным числом.

```python
# присваиваем столбцам parks_around_3000 и ponds_around_3000 значения с измененным типом данных
data['parks_around_3000'] = data['parks_around_3000'].astype('int64')

data['ponds_around_3000'] = data['ponds_around_3000'].astype('int64')
```

Теперь рассмотрим пропуски в столбцах **park_nearest** и **ponds_nearest**. Скорее всего, пропуски в этих столбцах связаны со значениями в соседних столбцах **parks_around_3000** и **ponds_around_3000** - парки и пруды в радиусе 3 км. Проверим это. Заполним пропуски в столбцах **park_nearest** и **ponds_nearest** нулями в тех строках, в которых значения столбцов **parks_around_3000** и **ponds_around_3000** также нулевые. Будем считать, что если парков и прудов нет в пределах 3 км, значит, поблизости их нет.

```python
# выберем только те записи, в которых значения столбцов parks_around_3000 и ponds_around_3000 нулевые,
# заполняем нулями соответствующие значения столбца parks_nearest
data['parks_nearest'] = data.query('parks_around_3000 == 0')['parks_nearest'].fillna(0) 

# заполняем нулями соответствующие значения столбца ponds_nearest 
data['ponds_nearest'] = data.query('ponds_around_3000 == 0')['ponds_nearest'].fillna(0)
```

Проверяем сколько пропусков осталось в этих столбцах.

```python
# ищем сумму пропущенных значений в столбце parks_nearest
print(f'Сумма пропусков в столбце parks_nearest - {data["parks_nearest"].isna().sum()}')

# ищем сумму пропущенных значений в столбце ponds_nearest
print(f'Сумма пропусков в столбце ponds_nearest - {data["ponds_nearest"].isna().sum()}')
```

У нас остались пустыми значения столбцов **parks_nearest** и **ponds_nearest** для которых в соответствующих столбцах **parks_around_3000** и **ponds_around_3000** значения не нулевые и не пустые. Т.к. значения в этих столбцах картографические, т.е. заполняются автоматически в зависимости от адреса квартиры, то, скорее всего, оставшиеся пропуски связаны с тем, что эти квартиры находятся в маленьких городах, где нет обозначенных парков или прудов. Логически заполнить эти значения не получится, да и они вряд ли понадобятся нам для дальнейшего анализа. Достаточно будет параметров **parks_around_3000** и **ponds_around_3000**. Поэтому сделаем эти значения нулевыми. И сразу же изменим тип данных на челочисленный, т.к. точность до сантиметров нам не нужна.

```python
#перезаписываем значение на значения с 0 вместо пропуска и меняем тип значений:
data['parks_nearest'] = data['parks_nearest'].fillna(0).astype('int64')

data['ponds_nearest'] = data['ponds_nearest'].fillna(0).astype('int64')
```

Проверяем, что в **parks_nearest** и в **ponds_nearest** нет пропусков.

Теперь рассмотрим параметры **airports_nearest** и **city_centers_nearest**. Начнем с **airports_nearest**. Найдем названия населенных пунктов, в которых есть непустые значения в этом столбце. Создадим список с названиями таких населенных пунктов.

```python
# считаем количество строк для каждого уникального значения столбца locality_name
# для которого выполняется условие, что значение столбца airports_nearest не пустое
data_airport_exist = (data.query('airports_nearest.isna() == False')['locality_name']
                          .value_counts())

data_airport_exist
```

Мы нашли список с названиями населенных пунктов, для которых хотя бы один раз указано расстояние до аэропорта. Можно отметить, что все эти населенные пункты находятся в каком-либо районе города Санкт-Петербурга, т.е. входят в его состав. Кроме Мурино и Щеглово, эти города относятся к ЛО. Для Мурино у нас есть всего два непустых значения, а всего строк с квартирами в этом городе - **590**. То же самое и для Щеглово - у нас есть всего одного непустое значение, а всего строк - **24**. Заполнять пустые значения для квартир в этих городах по такому количеству данных будет неверно. Теперь мы сможем заполнить пустые значения столбца **airports_nearest** медианными значениями расстояния до аэропорта для каждого населенного пункта отдельно, за исключением Мурино и Щеглово. Пропуски в этих столбцах корректнее заполнить нулями.

```python
#перебираем каждое уникальное значение столбца locality_name
for locality in data['locality_name'].unique():
    
    # проверяем условие, что текущее значение есть в списке с названиями населенных пунктов,
    # в которых есть данные о расстоянии до аэропорта
    # за исключением Мурино и Щеглово
    if locality in data_airport_exist and locality != ['Мурино', 'Щеглово']: 
        
        # выбираем строки, для которых значение в столбце locality_name равно текущему locality 
        # и в столбце airports_nearest пропуск
        data.loc[(data['locality_name'] == locality)&(data['airports_nearest'].isna()), 'airports_nearest'] = \
        data.loc[data['locality_name'] == locality, 'airports_nearest'].median() 
        # находим медиану значений столбца airports_nearest для строк, в которых значение столбца locality_name равно текущему
```

Тогда будем считать, что данный критерий существенно не повлияет на цену квартиры. Заполним пропуски в этих строках нулевыми значениями.

```python
# делаем срез с условием, что в столбце airports_nearest - пропуск, выбираем столбец locality_name
# и считаем кол-во уникальных значений
data.query('airports_nearest.isna() == True')['locality_name'].value_counts() 
```

```python
# заполняем пропуски в столбце airports_nearest нулями
data['airports_nearest'] = data['airports_nearest'].fillna(0)
```

Плюс ко всему, проанализировав, в каких населенных пунктах есть данные по расстоянию до центра города, можно отметить, что все эти населенные пункты находятся в каком-либо районе города Санкт-Петербурга, т.е. входят в его состав. Кроме Мурино и Щеглово. Для Мурино у нас есть всего два значения, причем одианковых, возможно, эти квартиры находятся в одном доме. По аналогии с удаленностью до аэропорта заполним пустые значения в столбце **city_centers_nearest** медианными для каждого города, только исключив из этого списка Мурино и Щеглово, т.к. заполнять пропуски в большом количестве строк значениями из пары строк не совсем корректно.

```python
# создадим таблицу, в которой будут строки, для которых известна общая площадь и площадь кухни:
data_area = good_data.query('total_area.isna() == False and kitchen_area.isna() == False')
```

Вероятнее, всего, для квартир или домов с большой площадью, доля кухни в общей площади будет меньше, чем для квартир с небольшой площадью. Выделим несколько категорий квартир. Первой будет категория с квартирами типичной площади - **до 50 кв.м**. Вторая категория **более 50 кв.м.**.

```python
# создадим таблицу и поместим в нее строки, для которых значение столбца total_area <= 50 кв.м.
data_area_first = data_area.query('total_area <= 50')

# найдем медиану отношения площади кухни к общей площади
first_median = (data_area_first['kitchen_area']/data_area_first['total_area']).median()

first_median
```

Получили значение около **0.22**. Проделаем то же самое для второй категории.

```python
# создадим таблицу и поместим в нее строки, для которых значение столбца total_area > 50 кв.м.:
data_area_second = data_area.query('total_area > 50')

# найдем медиану отношения площади кухни к общей площади:
second_median = (data_area_second['kitchen_area']/data_area_second['total_area']).median()

second_median
```

Как мы и думали, отношение площади кухни к общей площади для двух категорий квартир получилось разным. Теперь нам нужно найти строки, в которых известна общая площадь и неизвестна площадь кухни (для всех трех категорий). Так же нам необходимо исключить из этого списка квартиры-студии и квартиры со свободной планировкой, т.к. будем считать, что для них нет отдельного метража кухни.

```python
# выбираем столбец kitchen_area для строк, в которых количество комнат не нулевое,
# столбец studio имеет значение False
# столбец open_plan имеет значение False
# столбец total_area <= 50
# в столбце kitchen_area есть пропуск
good_data.loc[(good_data['total_area'] <= 50)& \
                      (good_data['rooms'] != 0)& \
                      (good_data['studio'] == False)& \
                      (good_data['open_plan'] == False)& \
                      (good_data['kitchen_area'].isna()), 'kitchen_area'] = \
                      round((good_data.loc[(good_data['total_area'] <= 50), 'total_area'])*first_median, 2)
                      # заполняем этот пропуск значением столбца total_area умноженным на медианный коэффициент для квартир первой группы

# выбираем столбец kitchen_area для строк, в которых количество комнат не нулевое,
# столбец studio имеет значение False
# столбец total_area > 50 кв.м.
# в столбце kitchen_area есть пропуск
good_data.loc[(good_data['total_area'] > 50)& \
              (good_data['rooms'] != 0)& \
              (good_data['studio'] == False)& \
              (good_data['kitchen_area'].isna()), 'kitchen_area'] = \
              round((good_data.loc[(good_data['total_area'] > 50), 'total_area'])*second_median, 2)
              # заполняем этот пропуск значением столбца total_area умноженным на медианный коэффициент для квартир второй группы
```

Теперь у нас остались пропуски в столбце **kitchen_area** только для квартир, у которых значение в столбце **rooms** нулевое и значение в столбце **studio** - True. Заполним оставшиеся пропуски нулями.

```python
# заполняем пропущенные столбца kitchen_area значения нулями
good_data['kitchen_area'] = good_data['kitchen_area'].fillna(0)
```

```python
# создадим таблицу со строками, в которых значение столбца total_area <= 50 кв.м.
data_living_first = data_living_area.query('total_area <= 50')

# найдем медиану отношения жилой площади к общей площади
living_median_first = (data_area_first['living_area']/data_area_first['total_area']).median()

living_median_first
```

```python
# создадим таблицу со строками, в которых значение столбца total_area > 50 кв.м.
data_living_second = data_living_area.query('total_area > 50')

# найдем медиану отношения жилой площади к общей площади
living_median_second = (data_area_second['living_area']/data_area_second['total_area']).median()

living_median_second
```

Медианы для трех групп не так уж сильно отличаются друг от друга. Теперь заполним пустые значения в столбце **living_area** с учетом найденного коэффициента.

```python
# выбираем столбец living_area для строк в которых,
# столбец total_area <= 50 кв.м.
# в столбце living_area есть пропуск

good_data.loc[(good_data['total_area'] <= 50)& \
              (good_data['living_area'].isna()), 'living_area'] = \
              round((good_data.loc[(good_data['total_area'] <= 50), 'total_area'])*living_median_first, 2) 
              # заполняем этот пропуск значением столбца total_area 
              # умноженным на медианный коэффициент для квартир первой группы

# выбираем столбец living_area для строк, в которых 
# столбец total_area > 50 кв.м
# в столбце living_area есть пропуск

good_data.loc[(good_data['total_area'] > 50)& \
              (good_data['living_area'].isna()), 'living_area'] = \
              round((good_data.loc[(good_data['total_area'] > 50), 'total_area'])*living_median_second, 2)
              # заполняем этот пропуск значением столбца total_area
              # умноженным на медианный коэффициент для квартир второй группы
```

```python
# создаем новый столбец year_exposition
# используем атрибут dt с методом year для вывода года
good_data['year_exposition'] = good_data['first_day_exposition'].dt.year
```

<a id='chapter_11'></a>Создадим еще один столбец с типом этажа квартиры ("первый", "последний" или "другой")

```python
# создаем новый столбец type_of_floor со значением первый для строк, в которых
# значение столбца floor равно 1 и не равно 0 (это пропуски в столбце floor)
good_data.loc[(good_data['floor'] == 1)&(good_data['floors_total'] != 0), 'type_of_floor'] = 'первый'

# создаем новый столбец type_of_floor со значением последний для строк, в которых
# значение столбца floor равно значению floors_total и не равно 0 (это пропуски в столбце floor)
good_data.loc[(good_data['floor'] == good_data['floors_total'])& \
              (good_data['floors_total'] != 0), 'type_of_floor'] = 'последний'

# создаем новый столбец type_of_floor со значением другой для строк, в которых
# значение столбца floor не равно значению floors_total и не равно 0 и 1
good_data.loc[(good_data['floor'] != good_data['floors_total'])& \
              (good_data['floor'] != 1)&(good_data['floors_total'] != 0), 'type_of_floor'] = 'другой'
```

Стоит отметить, что в новом столбце **type_of_floor** будут пропуски для строк, у которых нет значения **floors_total**.

<a id='chapter_12'></a>Еще одним столбцом станет расстояние до центра города в километрах - **city_centers_kms**.

```python
# создаем новый столбец разделив значения столбца city_centers_nearest на 1000 
# и переведем в целочисленный тип
good_data['city_centers_kms'] = round(good_data['city_centers_nearest']/1000).astype('int64')
```

```python
# создадим отдельную таблицу со строками только для Санкт-Петербурга
data_spb = good_data.copy().loc[good_data['locality_name'] == "Санкт-Петербург"]

# выводим получившуюся таблицу
data_spb.head()
```

Добавим в таблицу столбец с ценой за каждый километр от центра города.

```python
# создаем новый столбец и присваиваем ему значения общей цены/расстояние до центра в км
data_spb['price_meter_spb'] = data_spb['last_price']/data_spb['city_centers_kms']
```

Сгруппируем таблицу по столбцу расстояния до центра города и посчитаем медианную цену каждого километра.

```python
# группируем таблицу по столбцу city_centers_kms, считаем медиану по столбцу price_meter_spb 
# и сортируем значения по убыванию
price_per_meter_spb = (data_spb.groupby('city_centers_kms')['price_meter_spb']
                               .median()
                               .sort_values(ascending=False))

# строим столбчатую диаграмму по полученным значениям
price_per_meter_spb.plot(kind='bar',
                         title='Распределение цены за квадратный метр по удаленности от центра',
                         figsize=(8,6),
                         legend=False)

# задаем подпись оси x
plt.xlabel('Удаленность от центра, км')

# задаем подпись оси y
plt.ylabel('Цена за квадратный метр')

# выводим получившийся график
plt.show();
```
