Возможно, рейтинга пользователей и критиков нет у игр, которые были выпущены достаточно давно или же, наоборот, игра вышла недавно и оценки для нее попросту нет. В данном случае заполнить пропуски близкими к реальному значениями не получится, т.к. каждая строка в датафрейме - это отдельная игра со средними оценками, поэтому заполним пропуски в этом столбце нулями и учтем это при дальнейшем анализе. Так же изменим тип данных на целочисленный, но для начала проверим, что все значения в столбце целые, т.е. дробная часть числа только нули. 

```python
# находим строки, в которых значение critic_store не пустое 
# и остаток от деления на единицу не равен нулю
games.query('critic_score.isna() == False and critic_score%1 != 0')
```

Видим, что все числа в столбце - целые. Заполним пропуски нулями и изменим тип данных на целочисленный.

```python
# заменяем пропуски в столбце на нули и меняем тип данных на целочисленный
games['critic_score'] = games['critic_score'].fillna(0).astype(int)
```

Теперь рассмотрим столбец **user_score**. Выведем основные характеристики этого столбца.

```python
# выведем характеристики столбца user_score
games['user_score'].describe()
```

Видим, что непустых значений **9838**. Уникальных значений столбца - **96**, а так же есть значение **tbd**. Это расшифровывается как **to_be_determind**, скорее всего такой рейтинг у игр, для которых еще не определен пользовательский рейтинг, это могут быть новые игры или игры, для которых ранее не собиралась пользовательская оценка. По всей видимости, когда-то эти данные будут известны, но сейчас мы их логично заменить не можем. Таких строк у нас **2376** - это достаточно много. Заменим их так же на пропуски, т.к. заменять на нули нет необходимости - тип данных этого столбца должен быть вещественный.

```python
# заменяем значения tbd в столбце user_score на пустое значение
games.loc[games['user_score'] == 'tbd', 'user_score'] = np.NaN

# меняем тип данных на вещественный
games['user_score'] = games['user_score'].astype(float)
```

Проверим, что в столбце нет аномальных значений.

```python
# выведем основные статистические характеристики столбца
games['user_score'].describe()
```

Видим, что максимальное значение рейтинга - **9.7** (при максимальном - 10), но есть и минимальное нулевое значение. Выведем эти строки.

```python
games.query('user_score == 0')
```

Видим, что это одна строка, нулевое значение в столбце **critic_score** корректно будет заменить так же на пустое, т.к. нулевой рейтинг - это отсутствующий рейтинг.

```python
# заменяем нулевые значения в столбце user_score на пустое значение
games.loc[games['user_score'] == 0, 'user_score'] = np.NaN
```

Здесь дополнительно стоит отметить, что в столбце **critic_score** пропуски заполнены нулями, а в **user_score** пропуски заполнены пустым значением. Это необходимо будет учесть при дальнейшем анализе во избежание ошибок.

Теперь рассмотрим столбец **rating**. Выведем уникальные значения.

```python
# выводим уникальные значения столбца
games['rating'].unique()
```

Видим, что это столбец с возрастными категориями 'E', 'M', 'T', 'E10+', 'K-A', 'AO', 'EC', 'RP'. Так же в этом столбце есть пропуски, заполнить логично мы их не можем, поэтому заменим их на строковое значение 'Nan' для того, чтобы проще было анализировать в дальнейшем и не "потерять" их.

```python
# заполняем пропуски на строковое значение Nan
games['rating'] = games['rating'].fillna('Nan')
```

Снова выведем основные характеристики таблицы.

```python
#выводим основную информацию о данных
games.info()
```

## Часть 4. Проведем статистический анализ данных 
Проверим некоторые статистические гипотезы:
* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* Средние пользовательские рейтинги жанров Action и Sports разные.

Проверять гипотезы будем по данным за актульный период (т.е. с 2013 по 2016 гг.).

<a id='chapter_13'> </a>Сначала проверим равны ли пользовательские рейтинги платформ **XOne** и **PC**. Сформулируем нулевую гипотезу:

* H0: средний пользовательский рейтинг платформы **XOne** равен среднему пользовательскому рейтингу платформы **PC**

Тогда альтернативная гипотеза:

* H1: средний пользовательский рейтинг платформы **XOne** не равен среднему пользовательскому рейтингу платформы **PC**

Критический уровень статистической значимости примем равным 5%, тогда доверительный интервал будет равен 95%. То есть мы принимаем максимальную вероятность того, что нулевая гипотеза будет отвергнута при условии, что она верна - 5%, если полученное в результате теста значение окажется больше этого, значит, нулевая гипотеза не может быть отвергнута.   

Перед тем как выполнять статистический тест, проверим выборки на соответствие критериям, при которых можно применять t-тест:

* Генеральные совокупности не должны зависеть друг от друга.
* Выборочные средние должны быть нормально распределены. А если выборка всего одна, то её среднее должно иметь нормальное распределение для разных выборок этого размера из определённой генеральной совокупности.

Это условие не препятствие. Благодаря ЦПТ, если размер выборки составляет хотя бы несколько десятков значений, выборочные средние, которые можно получить из одной и той же генеральной совокупности, будут распределены нормально вокруг истинного среднего этой совокупности. И это утверждение верно, даже если сама генеральная совокупность не распределена нормально. 

Создадим отдельные массивы из значений пользовательского рейтинга для платформ PC и XOne.

```python
# создаем массив из значений пользовательского рейтинга для платформы XOne
xone_score = (platforms.loc[(platforms['platform']=="XOne")&\
                           (platforms['user_score'].isna()==False),'user_score'])

# выведем длину получившегося массива
len(xone_score)
```

```python
# создаем массив из значений пользовательского рейтинга для платформы PC
pc_score = (platforms.loc[(platforms['platform']=="PC")&\
                         (platforms['user_score'].isna()==False), 'user_score'])

# выведем длину получившегося массива
len(pc_score)
```

Итак, для корректного применения **t-test**, согласно центральной предельной теореме, говорящей нам о том, что среднее значение метрики по достаточно большой выборке пользователей будет иметь нормальное распределение, нам необходимо проверить, что распределение именно средних нормально. Если мы не знаем, можем ли считать распределение средних нормальным, то можем оценить это по одной нашей выборке, превратив ее во много выборок и посчитав среднее для всех. Так мы получим распределение, предполагая, что оно примерно совпадает с распределением среднего генеральной совокупности. 

Воспользуемся методом **Шапиро** для проверки нормальности распределения, предварительно создав списки, в которых будет по 1000 средних значений из 1000 выборок, сгенерированных на основе имеющихся данных. Нулевая гипотеза для Шапиро теста говорит о том, что случайная величина, выборка которой нам известна, распределена нормально. Уровень статистической значимости примем 0.05.

```python
# задаем уровень статистической значимости
alpha = .05

# создадим пустой список для средних значений выборок на основе пользовательских рейтингов для XOne
xone_list = []

# создадим пустой список для средних значений выборок на основе пользовательских рейтингов для PC
pc_list = []

# будем создавать 1000 выборок, находить для них среднее значение и заносить их в список
for i in range (1000):
    
    xone_list.append(xone_score.sample(1000, replace=True).mean())
    
    pc_list.append(pc_score.sample(1000, replace=True).mean())
    
# находим p-value для метода Шапиро    
result_xone = st.shapiro(xone_list).pvalue

result_pc = st.shapiro(pc_list).pvalue    

# строим гистограмму для полученных значений
sns.histplot(xone_list)

# задаем название графика
plt.title('Распределение среднего значения выборок, полученных из имеющейся выборки для XOne')

# выводим получившийся график
plt.show();

# сравниваем полученное значение p-value с заданным уровнем значимости
if result_xone > alpha:
    
    print('Распределение нормальное')
    
else:
    
    print('Распределение не нормальное, придется использовать другой статистический метод')

# строим гистограмму для полученных значений
sns.histplot(pc_list)

# задаем название графика
plt.title('Распределение среднего значения выборок, полученных из имеющейся выборки для PC')

# отображаем получившийся график
plt.show();

# сравниваем полученное значение p-value с заданным уровнем значимости
if result_pc > alpha:
    
    print('Распределение нормальное')
    
else:
    
    print('Распределение не нормальное, придется использовать другой статистический метод')
```

Видим, что оба получившихся распределения нормальны. Значит, мы можем использовать **t-test**.

```python
# обозначаем уровень статистической значимости
alpha = .05

# выполняем t-test для двух независимых выборок
# параметр equal_var=False, т.к. размеры выборок разные
results = st.ttest_ind(xone_score, pc_score, equal_var=False)

# выводим значение p-value
print('p-value равно', results.pvalue)

# проверяем равенство p-value и alpha
# т.к. гипотеза двусторонняя, то полученное pvalue пополам делить не нужно
if results.pvalue < alpha:
    
    print("Отвергаем нулевую гипотезу")
    
else:
    
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Сформулируем нулевую гипотезу:

* H0: средний пользовательский рейтинг игр жанра **Action** равен среднему пользовательскому рейтингу жанра **Sports**

Тогда альтернативная гипотеза:

* H1: средний пользовательский рейтинг игр жанра **Action** не равен среднему пользовательскому рейтингу жанра **Sports**

Критический уровень статистической значимости так же примем равным 5%.

```python
# создаем массив из значений пользовательского рейтинга для игр жанра Action
action_score = platforms.loc[(platforms['genre']=="Action")&(platforms['user_score'].isna()==False), 'user_score']

# создаем массив из значений пользовательского рейтинга для игр жанра Action
sports_score = platforms.loc[(platforms['genre']=="Sports")&(platforms['user_score'].isna()==False), 'user_score']

# выводим длину полученных массивов
len(action_score), len(sports_score)
```

Проверяем данные на нормальность аналогично предыдущей гипотезе.

```python
# задаем уровень статистической значимости
alpha = .05

# создадим пустой список для средних значений выборок на основе пользовательских рейтингов для Action
action_list = []

# создадим пустой список для средних значений выборок на основе пользовательских рейтингов для Sports
sports_list = []

# будем создавать 1000 выборок, находить для них среднее значение и заносить их в список
for i in range (1000):
    
    action_list.append(xone_score.sample(1000, replace=True).mean())
    
    sports_list.append(pc_score.sample(1000, replace=True).mean())
    
# находим p-value для метода Шапиро    
result_action = st.shapiro(action_list).pvalue

result_sports = st.shapiro(sports_list).pvalue    

# строим гистограмму для полученных значений
sns.histplot(action_list)

# задаем название графика
plt.title('Распределение среднего значения выборок, полученных из имеющейся выборки для Action')

# отражаем получившийся график
plt.show()

# сравниваем полученное значение p-value с заданным уровнем значимости
if result_action > alpha:
    
    print('Распределение нормальное')
    
else:
    
    print('Распределение не нормальное, придется использовать другой статистический метод')

# строим гистограмму для полученных значений
sns.histplot(sports_list)

# задаем название графика
plt.title('Распределение среднего значения выборок, полученных из имеющейся выборки для Sports')

# отражаем получившийся график
plt.show()

# сравниваем полученное значение p-value с заданным уровнем значимости
if result_sports > alpha:
    
    print('Распределение нормальное')
    
else:
    
    print('Распределение не нормальное, придется использовать другой статистический метод')
```

Видим, что оба распределения нормальные. Приступаем к проверке гипотезы.

```python
# обозначаем значение статистической значимости
alpha = .05

# выполняем t-test для двух независимых выборок
# параметр equal_var=False, т.к. размеры выборок разные
results2 = st.ttest_ind(action_score, sports_score, equal_var=False)

# выводим значение p-value
print('p-value равно', results2.pvalue)

# проверяем равенство p-value и alpha
# т.к. гипотеза двусторонняя, то полученное pvalue пополам делить не нужно
if results2.pvalue < alpha:
    
    print("Отвергаем нулевую гипотезу")
    
else:
    
    print("Не получилось отвергнуть нулевую гипотезу") 
```
